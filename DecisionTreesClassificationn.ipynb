# Observations
There are 150 rows and 5 columns
- There are no Null valuees
- There is one duplicated row
- The x-column ae sepal.length,sepal.width, petal.length and petal.width
- All the x-columns is "variety" which is categorical
- There are three flowers categories(classes)


```python
iris = iris.drop_duplicates(keep='first')
iris
```

```python
iris[iris.duplicated]
```

```python
iris = iris.reset_index(drop=True)
iris
```

# Perform labbel encoding of target column

```python
labelencoder = LabelEncoder()
iris.iloc[:,-1] = labelencoder.fit_transform(iris.iloc[:,-1])
iris.head()
```

```python
# Check the data types after label encoding
iris.info()
```

# Observation
- The target column ('varity') is still object type. it need to be converted to numeric(int)

```python
# Convert the target column data type to integer
iris['variety'] = pd.to_numeric(labelencoder.fit_transform(iris['variety']))
print(iris.info())
```

```python
# Divide the dataset in x-columns and y-columns
X=iris.iloc[:,0:4]
Y=iris['variety']
```

```python
x_train, x_test, y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
x_train
```

# Buliding Decision Tree Classifier Using Entropy Criteria

```python
model = DecisionTreeClassifier(criterion ='entropy',max_depth = None)
model.fit(x_train,y_train)
```

```python
# PLot the decision tree
plt.figure(dpi=1200)
tree.plot_tree(model);
```

```python
fn=['sepal length(cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']
plt.figure(dpi=1200)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True);
```
